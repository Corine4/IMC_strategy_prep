### Mid-price & Bid-ask spread

```python
# Integrate data
Data_total = pd.concat([Data0, Data1, Data2])

Data = Data_total[
    (Data_total['product'] == 'CHOCOLATE') |
    (Data_total['product'] == 'ROSES') |
    (Data_total['product'] == 'STRAWBERRIES') |
    (Data_total['product'] == 'GIFT_BASKET')
]

Data_N = Data.pivot_table(index='timestamp', columns='product', values='mid_price', aggfunc='mean')
F_Data = Data_N.dropna()
```

```python
# Comparison of mid-price trends of each product
fig, axs = plt.subplots(4, 1, figsize=(12, 8), sharex=True)
for i, col in enumerate(F_Data.columns):
    axs[i].plot(F_Data.index, F_Data[col], label=col)
    axs[i].set_ylabel(col)
    axs[i].legend()
    axs[i].grid()
plt.xlabel('Timestamp')
plt.suptitle('Mid Price Trend of Each Product')
plt.tight_layout()
plt.show()
```

```python
# Mid-price rise and fall direction
price_diff = F_Data.diff().apply(lambda x: x.map(lambda v: 1 if v > 0 else -1 if v < 0 else 0))
direction = price_diff.apply(pd.Series.value_counts)

print("Price Direction:")
print(direction)
```

### Basic Linear Regression

```python
print("Total Data information")
Data_total.head(10)
```

```python
# Reload data
Data_total = pd.concat([Data0, Data1, Data2])

Data = Data_total[
    (Data_total['product'] == 'CHOCOLATE') |
    (Data_total['product'] == 'ROSES') |
    (Data_total['product'] == 'STRAWBERRIES') |
    (Data_total['product'] == 'GIFT_BASKET')
]

Data_N = Data.pivot_table(index='timestamp', columns='product', values='mid_price', aggfunc='mean')
F_Data = Data_N.dropna()
```

```python
y = F_Data['GIFT_BASKET']
X = F_Data[['CHOCOLATE', 'ROSES', 'STRAWBERRIES']]
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
# Make predictions
y_pred = model.predict(X)

# Calculate residual
residual = y - y_pred

result = F_Data.copy()
result['Prediction'] = y_pred
result['Residual'] = residual

print("Linear Regression Result")
result.tail(10)
```

```python
plt.figure(figsize=(12, 3))
plt.plot(result.index, result['GIFT_BASKET'], label='Actual Price', color='blue')
plt.plot(result.index, result['Prediction'], label='Predicted Price', color='orange')
plt.title('Actual vs Predicted GIFT_BASKET Price')
plt.xlabel('Timestamp')
plt.ylabel('Mid Price')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 3))
plt.plot(result.index, result['Residual'], label='Residual', color='red')
plt.title('Residual (Actual - Predicted)')
plt.xlabel('Timestamp')
plt.ylabel('Residual')
plt.grid()
plt.show()
```

The actual price and the price predicted by CHOCOLATE, ROSES, and STRAWBERRIES regression are highly consistent, indicating that GIFT_BASKET can be approximately regarded as a combination of these three products;

The residual fluctuates between -100 and +100, providing possible deviation signals.

```python
# Calculation rolling correlation（windows can be changed）
rolling_window = 20
rolling_corr = pd.DataFrame(index=F_Data.index)

# Calculate the rolling correlation of each asset(products) with GIFT_BASKET separately
for asset in ['CHOCOLATE', 'ROSES', 'STRAWBERRIES']:
    rolling_corr[asset] = F_Data['GIFT_BASKET'].rolling(rolling_window).corr(F_Data[asset])

plt.figure(figsize=(12, 6))
for asset in ['CHOCOLATE', 'ROSES', 'STRAWBERRIES']:
    plt.plot(rolling_corr.index, rolling_corr[asset], label=f'Corr with {asset}')
plt.title(f'Rolling Correlation with GIFT_BASKET (window={rolling_window})')
plt.xlabel('Timestamp')
plt.ylabel('Correlation')
plt.axhline(y=0, linestyle='--', color='gray')
plt.legend()
plt.grid()
plt.show()
```

```python
# Residual: residual > 0, overestimation, residual < 0，underestimation
high_estimate_ratio = (result['Residual'] > 0).mean()
low_estimate_ratio = (result['Residual'] < 0).mean()

plt.figure(figsize=(10, 5))
plt.hist(result['Residual'], bins=50, color='steelblue', edgecolor='black')
plt.axvline(x=0, color='red', linestyle='--', label='Residual = 0')
plt.title('Distribution of Residuals: GIFT_BASKET vs Predicted Value')
plt.xlabel('Residual (Observed - Predicted)')
plt.ylabel('Frequency')
plt.legend()
plt.grid()
plt.show()
```

### Volatility

```python
GIFT_BASKET = Data_total[Data_total['product'] == 'GIFT_BASKET'].copy()
GIFT_BASKET = GIFT_BASKET.groupby('timestamp')['mid_price'].mean().to_frame()

# Calculate rolling volatility
GIFT_BASKET['rolling_vol'] = GIFT_BASKET['mid_price'].rolling(window=5).std()
print(GIFT_BASKET.head(10))
```

```python
plt.figure(figsize=(12, 4))
plt.plot(GIFT_BASKET.index, GIFT_BASKET['rolling_vol'], color='orange', label='Rolling Volatility (5-step)')
plt.title('GIFT_BASKET Rolling Volatility Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Rolling Std of Mid Price')
plt.grid()
plt.legend()
plt.tight_layout()
plt.show()
```

```python
vol_resid = GIFT_BASKET[['rolling_vol']].join(result[['Residual']]).dropna()

x = vol_resid['Residual']
y = vol_resid['rolling_vol']

x_fit = np.linspace(x.min(), x.max(), 200)

# Linear fitting (deg=1)
coeffs_linear = np.polyfit(x, y, deg=1)
linear_fit = np.poly1d(coeffs_linear)

y_fit_linear = linear_fit(x_fit)

plt.figure(figsize=(10, 5))
plt.scatter(x, y, alpha=0.5, label='Observed (Residual vs Rolling Volatility)')
plt.plot(x_fit, y_fit_linear, color='darkred', label=f'Linear Fit:\n{coeffs_linear[0]:.3e}·x + {coeffs_linear[1]:.3f}')
plt.axvline(x=0, linestyle='--', color='gray', label='mₜ = 0')
plt.xlabel('Residual')
plt.ylabel('Rolling Volatility')
plt.title('Linear Fit: Volatility ~ Residual')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

coeffs_linear
```

```python
# Constructing trading signal areas based on residual and volatility
# Calculate thresholds for residual and rolling_vol
resid_threshold = vol_resid['Residual'].abs().quantile(0.85)
vol_threshold = vol_resid['rolling_vol'].quantile(0.85)

# True when both residual and volatility are high
vol_resid['signal'] = (vol_resid['Residual'].abs() > resid_threshold) & (vol_resid['rolling_vol'] > vol_threshold)


plt.figure(figsize=(10, 5))
plt.scatter(vol_resid['Residual'], vol_resid['rolling_vol'], 
            c=vol_resid['signal'].map({True: 'red', False: 'skyblue'}), 
            alpha=0.5, label='Data Points')
plt.axhline(y=vol_threshold, color='gray', linestyle='--', label='Volatility Threshold')
plt.axvline(x=resid_threshold, color='gray', linestyle='--')
plt.axvline(x=-resid_threshold, color='gray', linestyle='--')
plt.xlabel('Residual')
plt.ylabel('Rolling Volatility')
plt.title('Trade Signal Zones Based on Residual & Volatility')
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()

print(vol_resid.head(10))
```

### Lag 

```python
GIFT_BASKET_Mid = GIFT_BASKET.set_index('timestamp')['mid_price']
lags = [1, 2, 3, 4, 5]
lag_corrs = {}
for lag in lags:
    shifted = GIFT_BASKET_Mid.shift(lag)
    lag_corrs[lag] = GIFT_BASKET_Mid.corr(shifted)
    
pd.DataFrame({
    'Lag': list(lag_corrs.keys()),
    'Correlation with Mid Price': list(lag_corrs.values())
}).set_index('Lag'), spread_vol_corr
```

The negative correlations between Lag 1 and 2 indicate a certain degree of reverse in prices in the short term.

The correlations between Lag 3–5 are close to 0 or slightly positive, indicating that there is no obvious linear dependence of prices over longer lags.

GIFT_BASKET’s mid_price lacks significant autocorrelation, and more complex or combined indicators should be considered.
